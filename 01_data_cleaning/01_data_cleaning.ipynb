# Cleaning & Encoding Churn Column

```python
#Cleaning Churn column
customers_df['Churn'] = customers_df['Churn'].replace({'Yes':1, 'No':0, 'Y': 1, 'N':0 , 'CHURNED':1, 'NO CHURN':0})
customers_df[customers_df['Churn'] == 'Unknown'].shape

#Dropping rows with 'Unknown' in Churn column since the churn status will be used to build model
customers_df = customers_df[customers_df['Churn'] != 'Unknown']
customers_df['Churn'] = customers_df['Churn'].astype(int)
customers_df.shape
```

## Cleaning TotalCharges column

```python
customers_df['TotalCharges'] = customers_df['TotalCharges'].replace(' ', np.nan)
customers_df['TotalCharges'] = customers_df['TotalCharges'].astype(str).str.replace(r'[^\d.]', '', regex=True)
customers_df['TotalCharges'] = pd.to_numeric(customers_df['TotalCharges'], errors='coerce')
customers_df['TotalCharges'].astype(float)
customers_df['TotalCharges'].fillna(0, inplace=True)
```
